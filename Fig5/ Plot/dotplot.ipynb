```python
import pandas as pd
```

```python
from plotnine import ggplot, geom_point, aes, scale_fill_distiller, theme_bw, geom_tile, theme, element_text, element_blank
from plotnine.facets import facet_grid
```

```python
plotting_df=pd.read_csv('./hu_tf_use_df.csv',header=0)
```

```python
plotting_df
```

```python
tmp = plotting_df[['index', 'TF', 'color_val']
        ].pivot_table(index = 'index', columns = 'TF'
        ).fillna(0)['color_val']
tmp
```

```python
idx_max = tmp.idxmax(axis = 0)
order = pd.concat([idx_max[idx_max == x] for x in tmp.index.tolist() if len(plotting_df[plotting_df == x]) > 0]).index.tolist()
plotting_df['TF'] = pd.Categorical(plotting_df['TF'], categories = order)

import plotnine
plotnine.options.figure_size = (15, 5)
```

```python
order
```

```python
from plotnine import ggplot, geom_point, aes, scale_fill_distiller, theme_bw, geom_tile, theme, element_text, element_blank
from plotnine.facets import facet_grid
```

```python
plotting_df
```

```python
plot = (
                ggplot(plotting_df, aes('TF', 'index'))
                + geom_tile(mapping = aes(fill = 'color_val'))
                + scale_fill_distiller(type = 'div', palette = 'RdYlBu')
                + geom_point(
                        mapping = aes(size = 'size_val'),
                        colour = "black")
                + theme(axis_title_x = element_blank(), axis_title_y = element_blank()))
```

```python
plot
```

```python
plot.save('./hg38_dotplot.pdf',limitsize=False)
```
